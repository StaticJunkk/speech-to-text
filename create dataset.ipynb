```python
import librosa as lb
import os
import json
```

```python
dataset_path = 'dataset'
json_path = 'data.json'
samples_taken = 22050 #worth 1 sec of sound
#n_fft to set a window for the fast fourier transform
def prep_dataset(dataset_path, json_path, n_mfcc =13, hop_length = 512, n_fft = 512 ):
    data = {
        "mappings":[],
        "labels":[],
        "MFCCs":[],
        "files":[]
    }
    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):
        #we need to put a condition so that we don't run it at root level
        if dirpath is not dataset_path:
            category = dirpath.split("/")[-1]
            data["mappings"].append(category)
            print(f"processing {category}")
        
        for f in filenames:
            file_path= os.path.join(dirpath, f)
            signal, sr = lb.load(file_path)
            if len(signal) >= samples_taken:
                signal = signal[:samples_taken]
                MFCCs = lb.feature.mfcc(signal, n_mfcc =n_mfcc, hop_length = hop_length, n_fft = n_fft)
                data["labels"].append(i-1)
                data["MFCCs"].append(MFCCs.T.tolist())
                data["files"].append(file_path)
                print(f" {file_path}:{i-1}")
        
            
    with open(json_path, "w") as fp:
        json.dump(data, fp, indent = 3)
    
        
```

```python
if __name__ == "__main__":
    prep_dataset(dataset_path, json_path)
```
